```
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Input,Conv2D,BatchNormalization,MaxPooling2D,Flatten,Dense,Dropout
from tensorflow.keras import datasets, layers, models



conv1 = Sequential(name="Convolution_block1")
conv1.add(Conv2D(4, kernel_size=3, padding='same', activation='relu'))
conv1.add(Conv2D(4, kernel_size=3, padding= 'same' ,activation= 'relu'))
#Convolution2D_2 = BatchNormalization()(Convolution2D_2)
conv1.add( MaxPooling2D())

conv2 = Sequential(name="Convolution_block2")
conv2.add(Conv2D(8, kernel_size=3, padding='same', activation='relu'))
conv2.add(Conv2D(8, kernel_size=3, padding='same', activation='relu'))
#Convolution2D_6 = BatchNormalization()(Convolution2D_6)
conv2.add(MaxPooling2D())

conv3 = Sequential(name="Convolution_block3")
conv3.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
conv3.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
conv3.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
#Convolution2D_11 = BatchNormalization()(Convolution2D_11)
conv3.add(MaxPooling2D())

conv4 = Sequential(name="Convolution_block4")
conv4.add(Conv2D(32, kernel_size=3, padding='same', activation='relu'))
conv4.add(Conv2D(32, kernel_size=3, padding='same', activation='relu'))
conv4.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
#Convolution2D_12 = BatchNormalization()(Convolution2D_12)
conv4.add(MaxPooling2D(name='MaxPooling2D_4'))

conv5 = Sequential(name="Convolution_block5")
conv5.add(Conv2D(32, kernel_size=3, padding='same', activation='relu'))
conv5.add(Conv2D(32, kernel_size=3, padding='same', activation='relu'))
conv5.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
#Convolution2D_16 = BatchNormalization()(Convolution2D_16)
conv5.add(MaxPooling2D(name='MaxPooling2D_5'))

denses = Sequential(name="Dense_Block")
denses.add(Flatten())
denses.add(Dense(512,activation= 'relu' ))
denses.add(Dropout(0.2))
denses.add(Dense(512,activation= 'relu' ))
denses.add(Dropout(0.2))
denses.add(Dense(1,activation= 'linear' ))

Input_1 = Input(shape=(64, 64, 3), name='Input_1')
x = conv1(Input_1)
x = conv2(x)
x = conv3(x)
x = conv4(x)
x = conv5(x)
output = denses(x)
model2 = Model([Input_1],[output])
```

```
from tensorflow.keras.utils import plot_model
plot_model(
    model2, to_file='model.png', show_shapes=True, show_layer_names=True,
    rankdir='TB', expand_nested=False, dpi=96
)
model2.summary()
```

```

model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mae'])

history = model.fit(X_train, y_train, epochs=50, 
                    validation_data=(X_test, y_test))
```
